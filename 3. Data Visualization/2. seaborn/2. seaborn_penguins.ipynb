# 🐧 Penguins Dataset - Seaborn & Pandas Exploration

Clean exercises + solutions for students

## 📌 Import Libraries

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
```

## 📌 Load Dataset

```
df = sns.load_dataset('penguins')
df.head()
```

## ❓ Q1: Check basic info

```
df.info()
```

## ❓ Q2: Check missing values

```
df.isnull().sum()
```

## ❓ Q3: Drop missing values

```
df_clean = df.dropna()
df_clean.head()
```

## ❓ Q4: Count penguins per species

```
df['species'].value_counts()
```

## ❓ Q5: Average body mass per species

```
df.groupby('species')['body_mass_g'].mean()
```

## 📊 Q6: Histogram of body mass

```
sns.histplot(df_clean['body_mass_g'], kde=True)
plt.show()
```

## 📊 Q7: Countplot species

```
sns.countplot(x='species', data=df)
plt.show()
```

## 📊 Q8: Boxplot body mass per species

```
sns.boxplot(x='species', y='body_mass_g', data=df_clean)
plt.show()
```

## 📊 Q9: Scatterplot flipper vs body mass

```
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=df_clean)
plt.show()
```

## 📊 Q10: Pairplot

```
sns.pairplot(df_clean, hue='species')
plt.show()
```
