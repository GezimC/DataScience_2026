## Dataset columns

The file contains football player statistics:

| Column | Meaning |
|---|---|
| `Player` | Player name |
| `Team` | National team |
| `Position` | Player position |
| `Goals` | Goals scored |
| `Assists` | Assists made |
| `Matches` | Matches played |
| `MinutesPlayed` | Total minutes played |
| `Rating` | Performance rating |

## Exercise 1 – Import Pandas and load the dataset

Load `kosova_team.xlsx` into a DataFrame called `df` and display the first 5 rows.

```python
import pandas as pd

df = pd.read_excel('kosova_team.xlsx')
df.head()
```

**Explanation:** `pd.read_excel()` reads the Excel file and creates a DataFrame. `head()` is used to quickly inspect the first rows.

## Exercise 2 – Basic dataset inspection

Display:
- number of rows and columns
- column names
- general DataFrame information

```python
print('Shape:', df.shape)
print('\nColumns:')
print(df.columns.tolist())

print('\nDataFrame info:')
df.info()
```

**Explanation:** `shape` shows dataset size, `columns` shows field names, and `info()` shows data types and missing-value counts.

## Exercise 3 – Check missing values and duplicates

Check if the dataset contains:
- missing values
- duplicated rows

```python
missing_values = df.isna().sum()
duplicated_rows = df.duplicated().sum()

print('Missing values per column:')
print(missing_values)

print('\nNumber of duplicated rows:', duplicated_rows)
```

**Explanation:** `isna().sum()` counts missing values per column. `duplicated().sum()` counts full duplicated rows.

## Exercise 4 – Select specific columns

Select only these columns and show the first 10 rows:

`Player`, `Team`, `Position`, `Rating`

```python
player_overview = df[['Player', 'Team', 'Position', 'Rating']]
player_overview.head(10)
```

**Explanation:** Double square brackets are used to select multiple columns from a DataFrame.

## Exercise 5 – Filter rows with one condition

Find all players who belong to **Kosovo**.

Display only:
`Player`, `Position`, `Goals`, `Assists`, `Rating`

```python
kosovo_players = df.loc[df['Team'] == 'Kosovo', ['Player', 'Position', 'Goals', 'Assists', 'Rating']]
kosovo_players
```

**Explanation:** `loc[row_condition, selected_columns]` filters rows and selects columns in one clean step.

## Exercise 6 – Filter rows using `isin()`

Find all players whose position is **Winger** or **Forward**.

```python
attackers = df[df['Position'].isin(['Winger', 'Forward'])]
attackers[['Player', 'Team', 'Position', 'Goals', 'Assists', 'Rating']]
```

**Explanation:** `isin()` is useful when a column can match one of several possible values.

## Exercise 7 – Filter with multiple conditions

Find players who:
- played at least 5 matches
- have a rating greater than 7.0

Display the result sorted by rating from highest to lowest.

```python
consistent_high_rated = df[(df['Matches'] >= 5) & (df['Rating'] > 7.0)]
consistent_high_rated.sort_values('Rating', ascending=False)
```

**Explanation:** Use `&` for AND conditions in Pandas. Each condition must be inside parentheses.

## Exercise 8 – Create a new calculated column

Create a new column called `GoalContribution`:

`GoalContribution = Goals + Assists`

Then display `Player`, `Team`, `Goals`, `Assists`, and `GoalContribution`.

```python
df['GoalContribution'] = df['Goals'] + df['Assists']
df[['Player', 'Team', 'Goals', 'Assists', 'GoalContribution']].head(10)
```

**Explanation:** A new column can be created by assigning a calculation to `df['NewColumn']`.

## Exercise 9 – Create a ratio column safely

Create `MinutesPerMatch`:

`MinutesPerMatch = MinutesPlayed / Matches`

Show the top 10 players with the highest minutes per match.

```python
df['MinutesPerMatch'] = df['MinutesPlayed'] / df['Matches']

df[['Player', 'Team', 'Position', 'Matches', 'MinutesPlayed', 'MinutesPerMatch']].sort_values('MinutesPerMatch', ascending=False).head(10)
```

**Explanation:** This creates a per-match metric, which is more comparable than total minutes alone.

## Exercise 10 – Sort data

Sort all players by:
1. `Rating` descending
2. `GoalContribution` descending

Show the top 8 players.

```python
top_players = df.sort_values(['Rating', 'GoalContribution'], ascending=[False, False])
top_players.head(8)
```

**Explanation:** `sort_values()` can sort by multiple columns. The `ascending` list controls each column separately.

## Exercise 11 – Find the best player by one column

Find the single player with the highest `MinutesPlayed`.

```python
most_minutes_player = df.loc[df['MinutesPlayed'].idxmax()]
most_minutes_player
```

**Explanation:** `idxmax()` returns the index of the highest value. `loc[]` then returns that full row.

## Exercise 12 – Count values by category

Count how many players there are in each `Position`.

```python
position_counts = df['Position'].value_counts()
position_counts
```

**Explanation:** `value_counts()` is used for frequency counts of categories.

## Exercise 13 – Average rating by team

Calculate the average `Rating` for each team.

```python
average_rating_by_team = df.groupby('Team')['Rating'].mean().sort_values(ascending=False)
average_rating_by_team
```

**Explanation:** `groupby()` splits the data by team, then `mean()` calculates the average rating for each group.

## Exercise 14 – Multiple aggregations by position

Group by `Position` and calculate:
- average goals
- average assists
- average rating
- total players

```python
position_summary = (
    df.groupby('Position')
      .agg(
          Players=('Player', 'count'),
          AvgGoals=('Goals', 'mean'),
          AvgAssists=('Assists', 'mean'),
          AvgRating=('Rating', 'mean')
      )
      .sort_values('AvgRating', ascending=False)
)

position_summary
```

**Explanation:** Named aggregation with `.agg()` creates a clean summary table with custom column names.

## Exercise 15 – Team summary table

Create a summary table with one row per team and these columns:
- number of players
- total goals
- total assists
- average rating
- total minutes played

```python
team_summary = (
    df.groupby('Team')
      .agg(
          Players=('Player', 'count'),
          TotalGoals=('Goals', 'sum'),
          TotalAssists=('Assists', 'sum'),
          AvgRating=('Rating', 'mean'),
          TotalMinutes=('MinutesPlayed', 'sum')
      )
      .sort_values('AvgRating', ascending=False)
)

team_summary
```

**Explanation:** This is a typical reporting-style summary using `groupby()` and multiple aggregations.

## Exercise 16 – String filtering

Find all players whose name contains **`Rash`**.

```python
rash_players = df[df['Player'].str.contains('Rash', case=False, na=False)]
rash_players[['Player', 'Team', 'Position', 'Goals', 'Assists', 'Rating']]
```

**Explanation:** `str.contains()` searches inside text columns. `case=False` makes the search case-insensitive.

## Exercise 17 – Use `loc` to create a category column

Create a new column called `StarPlayer`:
- `Yes` if `Rating >= 7.3`
- `No` otherwise

Display `Player`, `Team`, `Rating`, and `StarPlayer`.

```python
df['StarPlayer'] = 'No'
df.loc[df['Rating'] >= 7.3, 'StarPlayer'] = 'Yes'

df[['Player', 'Team', 'Rating', 'StarPlayer']].sort_values('Rating', ascending=False)
```

**Explanation:** `loc` can update values only where a condition is true.

## Exercise 18 – Use `apply()` for custom logic

Create a column called `PerformanceLevel` using this logic:
- `Excellent` if rating is 7.5 or higher
- `Good` if rating is from 7.0 to below 7.5
- `Average` if rating is below 7.0

```python
def classify_performance(rating):
    if rating >= 7.5:
        return 'Excellent'
    elif rating >= 7.0:
        return 'Good'
    else:
        return 'Average'


df['PerformanceLevel'] = df['Rating'].apply(classify_performance)
df[['Player', 'Team', 'Rating', 'PerformanceLevel']].sort_values('Rating', ascending=False)
```

**Explanation:** `apply()` is useful when the logic is easier to write as a Python function.

## Exercise 19 – Use `groupby()` with `transform()`

Create a new column called `TeamAvgRating` that contains each player's team average rating.

Then create `RatingDifferenceFromTeam`:

`RatingDifferenceFromTeam = Rating - TeamAvgRating`

Show the players with the highest positive difference.

```python
df['TeamAvgRating'] = df.groupby('Team')['Rating'].transform('mean')
df['RatingDifferenceFromTeam'] = df['Rating'] - df['TeamAvgRating']

rating_difference = df[['Player', 'Team', 'Rating', 'TeamAvgRating', 'RatingDifferenceFromTeam']]     .sort_values('RatingDifferenceFromTeam', ascending=False)

rating_difference.head(10)
```

**Explanation:** `transform()` returns a value for every original row, so it is perfect when we need group statistics inside the original DataFrame.

## Exercise 20 – Find players with assists but no goals

Find players who have:
- `Goals == 0`
- `Assists >= 1`

```python
assist_no_goal = df[(df['Goals'] == 0) & (df['Assists'] >= 1)]
assist_no_goal[['Player', 'Team', 'Position', 'Goals', 'Assists', 'Rating']]
```

**Explanation:** This combines two numeric conditions using `&`.

## Exercise 21 – Pivot table

Create a pivot table showing total goals by:
- rows: `Position`
- columns: `Team`

Fill missing values with 0.

```python
goals_pivot = pd.pivot_table(
    df,
    values='Goals',
    index='Position',
    columns='Team',
    aggfunc='sum',
    fill_value=0
)

goals_pivot
```

**Explanation:** A pivot table reshapes grouped results into a more report-friendly table.

## Exercise 22 – Compare teams by position

Create a table that shows the average rating for each combination of `Team` and `Position`.

```python
team_position_rating = (
    df.groupby(['Team', 'Position'])['Rating']
      .mean()
      .reset_index()
      .sort_values(['Team', 'Rating'], ascending=[True, False])
)

team_position_rating
```

**Explanation:** Grouping by multiple columns allows us to analyze combinations, not only one category at a time.

## Exercise 23 – Query method

Use `.query()` to find players from **Turkey** with rating at least **7.2**.

```python
turkey_top_players = df.query("Team == 'Turkey' and Rating >= 7.2")
turkey_top_players[['Player', 'Team', 'Position', 'Rating', 'GoalContribution']]
```

**Explanation:** `.query()` can make filtering expressions easier to read, especially for simple conditions.

Export the updated DataFrame to a new Excel file named:

`kosova_team_with_solutions.xlsx`

The exported file should include all new columns created in this notebook.

```python
df.to_excel('kosova_team_with_solutions.xlsx', index=False)
print('File exported successfully: kosova_team_with_solutions.xlsx')
```

**Explanation:** `to_excel()` saves the DataFrame as an Excel file. `index=False` avoids exporting the Pandas index as an extra column.
