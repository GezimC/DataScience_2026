## 10. Filter Electronics Transactions

**Request:**  
Find all transactions where `Category` is `Electronics`.

**Explanation:**  
A condition returns `True` or `False` for each row. Pandas keeps only rows where the condition is `True`.


```python
electronics = df[df["Category"] == "Electronics"]

electronics
```

## 11. Filter Transactions with Quantity Greater Than 1

**Request:**  
Find all transactions where `Quantity` is greater than 1.


```python
df[df["Quantity"] > 1]
```

## 12. Filter Using Multiple Conditions

**Request:**  
Find all transactions where:

- `Category` is `Fashion`
- `Quantity` is at least 2

**Explanation:**  
Use `&` for AND conditions. Each condition should be inside parentheses.


```python
fashion_quantity = df[
    (df["Category"] == "Fashion") &
    (df["Quantity"] >= 2)
]

fashion_quantity
```

## 13. Filter Using `isin`

**Request:**  
Find all transactions where the product is either `Laptop` or `Phone`.

**Explanation:**  
`isin()` is useful when checking if a value belongs to a list of allowed values.


```python
df[df["Product"].isin(["Laptop", "Phone"])]
```

## 14. Create a Total Column

**Request:**  
Create a new column named `Total`, calculated as:

`Quantity * Price`

**Explanation:**  
This creates the total value of each transaction.


```python
df["Total"] = df["Quantity"] * df["Price"]

df.head()
```

## 15. Create a Price Level Column

**Request:**  
Create a new column named `PriceLevel`.

Rules:

- `High` if `Price >= 500`
- `Low` otherwise

**Explanation:**  
This is a simple example of creating categories from numeric data.


```python
df["PriceLevel"] = df["Price"].apply(lambda price: "High" if price >= 500 else "Low")

df[["Product", "Price", "PriceLevel"]].head(10)
```

## 16. Total Revenue

**Request:**  
Calculate the total revenue from all transactions.

**Explanation:**  
After creating the `Total` column, the total revenue is the sum of that column.


```python
total_revenue = df["Total"].sum()

total_revenue
```

## 17. Average, Minimum, and Maximum Price

**Request:**  
Find:

- Average product price
- Minimum product price
- Maximum product price


```python
price_summary = {
    "Average Price": df["Price"].mean(),
    "Minimum Price": df["Price"].min(),
    "Maximum Price": df["Price"].max()
}

price_summary
```

## 18. Count Transactions by Category

**Request:**  
Find how many transactions belong to each category.

**Explanation:**  
`value_counts()` counts how many times each unique value appears.


```python
df["Category"].value_counts()
```

## 19. Count Unique Customers and Products

**Request:**  
Find the number of unique customers and unique products.

**Explanation:**  
`nunique()` counts distinct values.


```python
unique_summary = {
    "Unique Customers": df["Customer"].nunique(),
    "Unique Products": df["Product"].nunique()
}

unique_summary
```

## 20. Total Revenue by Customer

**Request:**  
Calculate total revenue for each customer.

**Explanation:**  
`groupby()` groups rows by customer, then `sum()` calculates the total per customer.


```python
revenue_by_customer = df.groupby("Customer")["Total"].sum()

revenue_by_customer
```

## 21. Total Quantity Sold by Product

**Request:**  
Calculate the total quantity sold for each product.


```python
quantity_by_product = df.groupby("Product")["Quantity"].sum()

quantity_by_product
```

## 22. Average Price by Category

**Request:**  
Find the average price for each category.


```python
avg_price_by_category = df.groupby("Category")["Price"].mean()

avg_price_by_category
```

## 23. Multiple Aggregations by Category

**Request:**  
For each category, calculate:

- Total quantity sold
- Total revenue
- Average price

**Explanation:**  
`agg()` allows us to calculate multiple summary statistics at once.


```python
category_summary = df.groupby("Category").agg(
    TotalQuantity=("Quantity", "sum"),
    TotalRevenue=("Total", "sum"),
    AveragePrice=("Price", "mean")
)

category_summary
```

## 24. Customer with the Highest Total Spending

**Request:**  
Find which customer spent the most money in total.

**Explanation:**  
First group by customer, then use `idxmax()` to find the customer with the highest value.


```python
customer_revenue = df.groupby("Customer")["Total"].sum()

top_customer = customer_revenue.idxmax()
top_customer_amount = customer_revenue.max()

top_customer, top_customer_amount
```

## 25. Product with the Highest Revenue

**Request:**  
Find which product generated the highest revenue.


```python
product_revenue = df.groupby("Product")["Total"].sum()

top_product = product_revenue.idxmax()
top_product_revenue = product_revenue.max()

top_product, top_product_revenue
```

## 26. Product Sold the Most by Quantity

**Request:**  
Find which product was sold the most based on total quantity.


```python
product_quantity = df.groupby("Product")["Quantity"].sum()

most_sold_product = product_quantity.idxmax()
most_sold_quantity = product_quantity.max()

most_sold_product, most_sold_quantity
```

## 27. Sort Transactions by Total Value

Sort the DataFrame by `Total` from highest to lowest.


```python
df.sort_values(by="Total", ascending=False)
```

## 28. Top 5 Transactions by Total Value

Display the 5 transactions with the highest `Total`.

## 29. Transactions Above Average Total

**Request:**  
Find all transactions where `Total` is greater than the average transaction total.


```python
average_total = df["Total"].mean()

df[df["Total"] > average_total]
```

## 30. Add Customer Total Spending Using `transform`

**Request:**  
Create a new column named `CustomerTotalSpending` that shows each customer's total spending on every row.

**Explanation:**  
`transform()` returns values with the same length as the original DataFrame, so it is useful when we want group results added back to each row.


```python
df["CustomerTotalSpending"] = df.groupby("Customer")["Total"].transform("sum")

df[["Customer", "Product", "Total", "CustomerTotalSpending"]].head(10)
```

## 31. Calculate Each Transaction's Share of Customer Spending

**Request:**  
Create a new column named `ShareOfCustomerSpending`.

Formula:

`Total / CustomerTotalSpending`

**Explanation:**  
This shows how much each transaction contributed to the customer's total spending.


```python
df["ShareOfCustomerSpending"] = df["Total"] / df["CustomerTotalSpending"]

df[["Customer", "Product", "Total", "CustomerTotalSpending", "ShareOfCustomerSpending"]].head(10)
```

## 32. Pivot Table: Revenue by Customer and Category

**Request:**  
Create a pivot table showing total revenue by `Customer` and `Category`.

**Explanation:**  
A pivot table summarizes data in a matrix format.


```python
revenue_pivot = pd.pivot_table(
    df,
    values="Total",
    index="Customer",
    columns="Category",
    aggfunc="sum",
    fill_value=0
)

revenue_pivot
```

## 33. Customers Who Bought from Both Categories

**Request:**  
Find customers who bought products from both `Electronics` and `Fashion`.

**Explanation:**  
Group by customer, collect unique categories, then check which customers have both categories.


```python
customer_categories = df.groupby("Customer")["Category"].unique()

customers_both_categories = customer_categories[
    customer_categories.apply(lambda categories: {"Electronics", "Fashion"}.issubset(set(categories)))
]

customers_both_categories
```

## 34. Use `query()` for Cleaner Filtering

**Request:**  
Find all transactions where:

- `Category` is `Electronics`
- `Price` is at least 500
- `Quantity` is at least 2

**Explanation:**  
`query()` can make filtering more readable, especially when conditions are simple.


```python
df.query("Category == 'Electronics' and Price >= 500 and Quantity >= 2")
```

## 35. Final Business Summary Table

**Request:**  
Create a summary table by customer with:

- Number of transactions
- Total quantity
- Total spending
- Average transaction value

Sort the result by total spending descending.

**Explanation:**  
This combines grouping, aggregation, and sorting.


```python
customer_summary = df.groupby("Customer").agg(
    NumberOfTransactions=("TransactionID", "count"),
    TotalQuantity=("Quantity", "sum"),
    TotalSpending=("Total", "sum"),
    AverageTransactionValue=("Total", "mean")
).sort_values(by="TotalSpending", ascending=False)

customer_summary
```

## 36. Export Results to Excel

**Request:**  
Export the cleaned dataset and summary tables to a new Excel file named `transactions_analysis_solution.xlsx`.

**Explanation:**  
`ExcelWriter` allows us to save multiple DataFrames into different sheets of the same Excel file.


```python
with pd.ExcelWriter("transactions_analysis_solution.xlsx") as writer:
    df.to_excel(writer, sheet_name="Cleaned_Data", index=False)
    category_summary.to_excel(writer, sheet_name="Category_Summary")
    customer_summary.to_excel(writer, sheet_name="Customer_Summary")
    revenue_pivot.to_excel(writer, sheet_name="Revenue_Pivot")

print("Excel file exported successfully.")
```
