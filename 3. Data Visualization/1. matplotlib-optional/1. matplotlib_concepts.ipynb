## 1. Import Libraries

### Request
Import the libraries needed for this notebook.

### Explanation
We use:

- `matplotlib.pyplot` for plotting
- `numpy` for creating numerical data
- `pandas` for working with tabular data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 2. Basic Line Plot

### Request
Create a simple line plot using two lists.

### Explanation
A line plot is useful when we want to show how one value changes compared to another.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

## 3. Add Title and Axis Labels

### Request
Add a title, x-axis label, and y-axis label to the line plot.

### Explanation
Good charts should always explain what the axes represent.

```python
plt.plot(x, y)

plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

## 4. Change Line Style, Marker, and Width

### Request
Customize the line using markers, line style, and line width.

### Explanation
Matplotlib allows us to style charts to make them easier to read.

```python
plt.plot(x, y, marker="o", linestyle="--", linewidth=2)

plt.title("Customized Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

## 5. Create Data with NumPy

### Request
Create values from 0 to 10 and plot `y = x²`.

### Explanation
NumPy helps us create numerical sequences easily.

```python
x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y)

plt.title("y = x²")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## 6. Multiple Lines on the Same Plot

### Request
Plot `x`, `x²`, and `x³` on the same figure.

### Explanation
Multiple lines help us compare different trends.

```python
x = np.linspace(0, 5, 100)

plt.plot(x, x, label="x")
plt.plot(x, x**2, label="x²")
plt.plot(x, x**3, label="x³")

plt.title("Multiple Lines")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

## 7. Figure Size

### Request
Change the size of the figure.

### Explanation
`figsize=(width, height)` controls the size of the chart.

```python
plt.figure(figsize=(8, 4))

plt.plot(x, x**2)

plt.title("Figure Size Example")
plt.xlabel("x")
plt.ylabel("x²")

plt.show()
```

## 8. Grid Lines

### Request
Add grid lines to the chart.

### Explanation
Grid lines make it easier to read values from the chart.

```python
plt.plot(x, x**2)

plt.title("Plot with Grid")
plt.xlabel("x")
plt.ylabel("x²")
plt.grid(True)

plt.show()
```

# Part A — Object-Oriented Matplotlib

Matplotlib can be used in two styles:

1. `plt.plot()` style
2. Object-oriented style using `fig, ax = plt.subplots()`

The object-oriented style is better for real projects.

## 9. Create Figure and Axes

### Request
Create a plot using `fig, ax = plt.subplots()`.

### Explanation
`fig` is the whole figure.
`ax` is the actual plotting area.

```python
fig, ax = plt.subplots()

ax.plot(x, x**2)

ax.set_title("Object-Oriented Plot")
ax.set_xlabel("x")
ax.set_ylabel("x²")

plt.show()
```

## 10. Customize Axes with Object-Oriented Style

### Request
Add title, labels, legend, and grid using `ax`.

### Explanation
This style is cleaner when charts become more complex.

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, x, label="x")
ax.plot(x, x**2, label="x²")

ax.set_title("Object-Oriented Customization")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid(True)

plt.show()
```

# Part B — Common Chart Types

## 11. Scatter Plot

### Request
Create a scatter plot showing the relationship between two variables.

### Explanation
Scatter plots are used to show relationships between numerical variables.

```python
np.random.seed(42)

study_hours = np.random.randint(1, 10, 30)
exam_scores = study_hours * 8 + np.random.randint(40, 70, 30)

plt.scatter(study_hours, exam_scores)

plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")

plt.show()
```

## 12. Bar Chart

### Request
Create a bar chart showing sales by product.

### Explanation
Bar charts are useful for comparing categories.

```python
products = ["Laptop", "Phone", "Tablet", "Monitor"]
sales = [120, 250, 90, 160]

plt.bar(products, sales)

plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.show()
```

## 13. Horizontal Bar Chart

### Request
Create a horizontal bar chart.

### Explanation
Horizontal bar charts are useful when category names are long.

```python
plt.barh(products, sales)

plt.title("Sales by Product")
plt.xlabel("Sales")
plt.ylabel("Product")

plt.show()
```

## 14. Histogram

### Request
Create a histogram of exam scores.

### Explanation
Histograms show the distribution of numerical data.

```python
scores = np.random.normal(loc=75, scale=10, size=100)

plt.hist(scores, bins=10)

plt.title("Distribution of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```

## 15. Box Plot

### Request
Create a box plot of exam scores.

### Explanation
Box plots show median, spread, and possible outliers.

```python
plt.boxplot(scores)

plt.title("Box Plot of Exam Scores")
plt.ylabel("Score")

plt.show()
```

## 16. Pie Chart

### Request
Create a pie chart showing market share.

### Explanation
Pie charts show parts of a whole.
Use them only when there are a small number of categories.

```python
labels = ["Product A", "Product B", "Product C", "Product D"]
market_share = [35, 30, 20, 15]

plt.pie(market_share, labels=labels, autopct="%1.1f%%")

plt.title("Market Share")

plt.show()
```

# Part C — Subplots

## 17. Two Subplots Side by Side

### Request
Create two plots in one figure.

### Explanation
Subplots allow us to compare multiple charts in the same figure.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, x**2)
axes[0].set_title("x²")

axes[1].plot(x, x**3)
axes[1].set_title("x³")

plt.tight_layout()
plt.show()
```

## 18. Four Subplots

### Request
Create a 2x2 grid of plots.

### Explanation
This is useful for dashboards or comparing several variables.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, x)
axes[0, 0].set_title("x")

axes[0, 1].plot(x, x**2)
axes[0, 1].set_title("x²")

axes[1, 0].plot(x, np.sqrt(x))
axes[1, 0].set_title("sqrt(x)")

axes[1, 1].plot(x, np.sin(x))
axes[1, 1].set_title("sin(x)")

plt.tight_layout()
plt.show()
```

# Part D — Matplotlib with Pandas

## 19. Create a Small DataFrame

### Request
Create a DataFrame with monthly sales data.

### Explanation
In real data analysis, we usually plot data from a DataFrame.

```python
sales_df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [1200, 1500, 1800, 1600, 2100, 2400],
    "expenses": [800, 900, 1100, 1000, 1300, 1500]
})

sales_df
```

## 20. Line Plot from a DataFrame

### Request
Plot monthly sales using Matplotlib.

### Explanation
We use DataFrame columns as x and y values.

```python
plt.plot(sales_df["month"], sales_df["sales"], marker="o")

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

## 21. Compare Sales and Expenses

### Request
Plot sales and expenses in the same chart.

### Explanation
This helps us compare income and cost trends.

```python
plt.plot(sales_df["month"], sales_df["sales"], marker="o", label="Sales")
plt.plot(sales_df["month"], sales_df["expenses"], marker="o", label="Expenses")

plt.title("Sales vs Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.grid(True)

plt.show()
```

## 22. Add a Profit Column

### Request
Create a new column called `profit`.

### Explanation
Profit is calculated as:

`profit = sales - expenses`

```python
sales_df["profit"] = sales_df["sales"] - sales_df["expenses"]

sales_df
```

## 23. Bar Chart of Profit

### Request
Create a bar chart showing monthly profit.

### Explanation
A bar chart is good for comparing profit month by month.

```python
plt.bar(sales_df["month"], sales_df["profit"])

plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")

plt.show()
```

## 24. Highlight Highest Profit Month

### Request
Find the month with the highest profit.

### Explanation
Before visualizing or reporting results, we often calculate key insights.

```python
highest_profit_row = sales_df.loc[sales_df["profit"].idxmax()]

highest_profit_row
```

## 25. Annotate a Chart

### Request
Create a profit bar chart and annotate the highest profit month.

### Explanation
Annotations help highlight important values directly on a chart.

```python
plt.figure(figsize=(8, 4))

plt.bar(sales_df["month"], sales_df["profit"])

max_month = highest_profit_row["month"]
max_profit = highest_profit_row["profit"]

plt.annotate(
    f"Highest: {max_profit}",
    xy=(max_month, max_profit),
    xytext=(max_month, max_profit + 100),
    arrowprops=dict(arrowstyle="->")
)

plt.title("Monthly Profit with Annotation")
plt.xlabel("Month")
plt.ylabel("Profit")

plt.show()
```

# Part E — Useful Customizations

## 26. Rotate X-Axis Labels

### Request
Rotate x-axis labels.

### Explanation
This is useful when labels overlap.

```python
long_products = ["Gaming Laptop", "Smart Phone Pro", "Wireless Headphones", "External Monitor"]
sales = [120, 250, 180, 160]

plt.bar(long_products, sales)

plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=30)

plt.show()
```

## 27. Set Axis Limits

### Request
Control the visible range of the y-axis.

### Explanation
Axis limits can focus attention on a specific value range.

```python
plt.plot(sales_df["month"], sales_df["sales"], marker="o")

plt.title("Monthly Sales with Y-Axis Limit")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.ylim(1000, 2600)

plt.show()
```

## 28. Save a Figure

### Request
Save a chart as a PNG file.

### Explanation
`plt.savefig()` saves the current figure to a file.
Use it before or together with `plt.show()`.

```python
plt.plot(sales_df["month"], sales_df["profit"], marker="o")

plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")

plt.savefig("monthly_profit.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Final Summary

## Key Takeaways

- `plt.plot()` is used for line charts.
- `plt.scatter()` is used for relationships between numerical variables.
- `plt.bar()` and `plt.barh()` compare categories.
- `plt.hist()` shows distributions.
- `plt.boxplot()` shows spread and outliers.
- `fig, ax = plt.subplots()` is the recommended style for more complex charts.
- Always use titles, labels, and legends when needed.
- Use `plt.tight_layout()` when using subplots.
- Use `plt.savefig()` to export charts.
